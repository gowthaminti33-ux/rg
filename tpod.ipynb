```python
def video_to_frames(video_path: str, frames_path: str, size=(960, 540), start_index=0, ext=".jpg"):
    os.makedirs(frames_path, exist_ok=True)

    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        raise RuntimeError(f"❌ Cannot open video: {video_path}")

    frame_id = start_index
    saved = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame = cv2.resize(frame, size, interpolation=cv2.INTER_AREA)
        out_path = os.path.join(frames_path, f"{frame_id}{ext}")
        cv2.imwrite(out_path, frame)
        frame_id += 1
        saved += 1
    cap.release()
    print(f"✅ Extracted {saved} frames → {frames_path}")
```

```python
import os
import shutil
import xml.etree.ElementTree as ET

VIDEOS_ROOT = "videos/02_ashlee"
NEW_FRAMES_ROOT = "new_frames/02_ashlee"

def get_min_max_frame_from_xml(xml_path):
    tree = ET.parse(xml_path)
    root = tree.getroot()

    frames = []

    for polygon in root.iter('polygon'):
        frame = polygon.attrib.get('frame')
        if frame is not None:
            frames.append(int(frame))

    if not frames:
        return None, None

    return min(frames), max(frames)


def move_frames(relative_path, min_frame, max_frame):
    source_frames_dir = os.path.join(NEW_FRAMES_ROOT, relative_path, "frames")
    target_frames_dir = os.path.join(VIDEOS_ROOT, relative_path, "frames")

    if not os.path.exists(source_frames_dir):
        print(f"❌ Missing source: {source_frames_dir}")
        return

    os.makedirs(target_frames_dir, exist_ok=True)

    for file in sorted(os.listdir(source_frames_dir)):
        if not file.lower().endswith(".jpg"):
            continue

        frame_no = int(os.path.splitext(file)[0])

        # ✅ Move only frames between min and max (inclusive)
        if min_frame <= frame_no <= max_frame:
            src = os.path.join(source_frames_dir, file)
            dst = os.path.join(target_frames_dir, file)
            shutil.move(src, dst)
            print(f"✅ Moved {file} → {target_frames_dir}")


for root, dirs, files in os.walk(VIDEOS_ROOT):
    if "annotations.xml" in files:
        xml_path = os.path.join(root, "annotations.xml")
        min_frame, max_frame = get_min_max_frame_from_xml(xml_path)

        if min_frame is None:
            print(f"⚠️ No frame data in {xml_path}")
            continue

        # Remove /annotations from path and make relative to VIDEOS_ROOT
        relative_path = os.path.relpath(
            root.replace("/annotations", ""),
            VIDEOS_ROOT
        )
        move_frames(relative_path, min_frame, max_frame)
```

```python
import xml.etree.ElementTree as ET
import os
import shutil
from pathlib import Path
import glob

# ============================================================
# CONFIG
# ============================================================

LABELS_ORDER = [
    "playerAcard1",
    "playerBcard1",
    "playerAcard2",
    "playerBcard2",
    "playerAcard3",  # anchor for final range
]

FOLDER_MAP = {
    "playerAcard1": "frames_pAc1",
    "playerBcard1": "frames_pBc1",
    "playerAcard2": "frames_pAc2",
    "playerBcard2": "frames_pBc2",
    "playerAcard3": "frames_pAc3",  # final bucket
}

IMAGE_EXTS = {".jpg", ".png"}


# ============================================================
# XML PARSING
# ============================================================

def get_label_start_frames(xml_path, target_labels):
    """
    Returns earliest visible frame (outside=0)
    across ALL tracks of each label.
    """
    tree = ET.parse(xml_path)
    root = tree.getroot()

    label_start = {}

    for track in root.iter("track"):
        label = track.attrib.get("label")
        if label not in target_labels:
            continue

        for poly in track:
            if "frame" not in poly.attrib:
                continue

            if int(poly.attrib.get("outside", 0)) == 0:
                frame = int(poly.attrib["frame"])

                if label not in label_start:
                    label_start[label] = frame
                else:
                    label_start[label] = min(label_start[label], frame)

    return label_start




# ============================================================
# RANGE BUILDER
# ============================================================

def build_ranges(label_starts):
    """
    Produces ordered (label, start, end) ranges.
    end is exclusive.
    """
    present_labels = [l for l in LABELS_ORDER if l in label_starts]

    ranges = []
    for i, label in enumerate(present_labels):
        start = label_starts[label]
        end = (
            label_starts[present_labels[i + 1]]
            if i + 1 < len(present_labels)
            else None
        )
        ranges.append((label, start, end))

    return ranges


# ============================================================
# CORE LOGIC
# ============================================================

def split_frames_by_labels(frames_dir, xml_path):
    frames_dir = Path(frames_dir)
    base_dir = frames_dir.parent

    label_starts = get_label_start_frames(xml_path, LABELS_ORDER)
    if not label_starts:
        print("❌ No labels found in XML:", xml_path)
        return

    ranges = build_ranges(label_starts)

    # Index all frames ONCE
    frames = []
    for f in frames_dir.iterdir():
        if f.suffix.lower() in IMAGE_EXTS:
            frames.append((int(f.stem), f))

    frames.sort(key=lambda x: x[0])
    moved = set()

    # Move frames per range
    for label, start, end in ranges:
        target_folder = FOLDER_MAP.get(label)
        if not target_folder:
            continue

        target_dir = base_dir / target_folder
        target_dir.mkdir(exist_ok=True)

        for frame_num, frame_file in frames:
            if frame_file in moved:
                continue

            if frame_num >= start and (end is None or frame_num < end):
                shutil.move(str(frame_file), target_dir / frame_file.name)
                moved.add(frame_file)

    # Move leftovers → frames_pAc3
    final_dir = base_dir / "frames_pAc3"
    final_dir.mkdir(exist_ok=True)

    for _, frame_file in frames:
        if frame_file not in moved and frame_file.exists():
            shutil.move(str(frame_file), final_dir / frame_file.name)
```

```python
dealer_paths = glob.glob("videos/01_jameica/**/frames", recursive=True)

for dealer_path in dealer_paths:
    parts = Path(dealer_path).parts

    _, dealer, scenario, folder, _ = parts[-5:]

    if scenario not in {"arrival", "front"}:
        continue

    if scenario == "front" and folder == "08":
        xml_path = f"videos/{dealer}/{scenario}/{folder}/annotations/annotations.xml"

        if not os.path.exists(xml_path):
            print("⚠️ Missing XML:", xml_path)
            continue

        print("✅ Processing:", dealer_path)
        split_frames_by_labels(dealer_path, xml_path)
```
